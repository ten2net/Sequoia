```python
import pandas as pd
import numpy as np
import akshare as ak
from datetime import datetime, timedelta
import time
import schedule
from tqdm import tqdm
from termcolor import colored
import argparse
import pywencai
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from joblib import dump, load
import pickle
```

```python
feature_columns = [
    '开盘'	,'开盘2分钟价格'	,'开盘2分钟涨幅',
    # 'open30', 
    'price25', 'volume25', 
    'close31', 'high31', 'low31', 'volume31', 'price31',
    'close32', 'high32', 'low32', 'volume32', 'price32', 
    '昨天开盘', '昨天最高', '昨天最低',
    '昨天收盘', '昨天成交量', '昨天涨跌幅', '昨天换手率'
    , '前天成交量', '前天涨跌幅', '前天换手率',
    '5日均价', '10日均价', '5日均量', '10日均量', '5日均换', '10日均换'
]

print("================未知数据==============")
file_path = f"""./results/甘州图灵甄选932_20240708_full-222.csv"""
df = pd.read_csv(file_path,dtype={'代码': 'str'})
df = df.dropna(subset=feature_columns)

model = load('./models/甘州图灵932_model.joblib')
scaler = load('./models/甘州图灵932_scaler.joblib')
data = scaler.transform(df[feature_columns])
predictions = model.predict(data)

df['预测排名'] = predictions 
print(df[:10][['浮盈排名' , '预测排名']])

print("================部分数据参与了训练==============")
file_path = f"""./results/甘州图灵甄选932_20240709_09-52_full.csv"""
df = pd.read_csv(file_path,dtype={'代码': 'str'})
df = df.dropna(subset=feature_columns)

with open('./models/甘州图灵932_model.pkl', 'rb') as file:
    model = pickle.load(file)
with open('./models/甘州图灵932_scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
data = scaler.transform(df[feature_columns])
predictions = model.predict(data)

df['预测排名'] = predictions 
print(df[:10][['浮盈排名' , '预测排名']])
```
