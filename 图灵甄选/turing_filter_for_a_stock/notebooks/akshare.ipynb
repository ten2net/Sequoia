```python
# '火箭发射', '快速反弹'

import akshare as ak
df = ak.stock_changes_em(symbol="快速反弹")
df=df[["时间","代码","名称","相关信息"]]
df.rename(columns={"时间":"time","代码":"code","名称":"name","相关信息":"info"},inplace=True)
# print(df[:50])
df =df[~(df['code'].apply(str).str.startswith('8')) & 
       ~(df['code'].apply(str).str.startswith('4')) &
       ~(df['name'].apply(str).str.startswith('ST')) & 
       ~(df['name'].apply(str).str.startswith('*'))  & 
       ~(df['name'].apply(str).str.startswith('N')) & 
       ~(df['name'].apply(str).str.startswith('C'))       
      ]
df[['volume','price','diff']] =df['info'].astype(str).str.split(",",expand=True).astype(float)
df['diff']=df['diff'].astype(float).abs()
df = df.query('price > 5 ')
df.sort_values(by='diff', ascending=False, inplace=True)
df[:-50]
```

```python

from pool.pool import AmountStockPool


def compute_market_sentiment_index(self) -> float:
    stockPool = AmountStockPool()
    symbols =stockPool.get_symbols(cloumn_name="amount", k=100)
    df = stockPool.get_data(symbols)
    # 计算加权市场情绪因子
    sentiment_index = (df['market_sentiment_factor'] * df['amount']).sum() / df['amount'].sum()

    print(f"加权市场情绪因子: {sentiment_index}")
    return sentiment_index
```
