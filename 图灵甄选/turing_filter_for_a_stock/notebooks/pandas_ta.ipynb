```python
class BollingerOBVStrategy:
    def __init__(self, data, window=14, num_std=2):
        self.data = data
        self.window = window
        self.num_std = num_std
        self.obv = self.calculate_obv()
        self.bollinger_bands = self.calculate_bollinger_bands()
        self.signals = self.generate_signals()

    def calculate_obv(self):
        # 计算OBV
        self.data['OBV'] = self.data['close'] * np.sign(self.data['volume'].diff())
        self.data['OBV'] = self.data['OBV'].cumsum()
        return self.data['OBV']

    def calculate_bollinger_bands(self):
        # 计算布林带
        middle_band = self.data['close'].rolling(window=self.window).mean()
        upper_band = middle_band + self.num_std * self.data['close'].rolling(window=self.window).std()
        lower_band = middle_band - self.num_std * self.data['close'].rolling(window=self.window).std()
        bollinger_bands = {'Middle': middle_band, 'Upper': upper_band, 'Lower': lower_band}
        self.data = pd.concat([self.data, pd.DataFrame(bollinger_bands)], axis=1)
        return bollinger_bands

    def generate_signals(self):
        # 生成交易信号
        signals = pd.DataFrame(index=self.data.index)
        signals =self.data.copy()
        signals['signal2'] = 0  # 0 for no signal, 1 for buy, -1 for sell

        # 检查OBV和布林带的交叉情况
        for i in range(1, len(self.data)):
            if (self.data['close'][i] > self.data['Upper'][i]) and (self.obv[i] < self.obv[i - 1]):
                signals.at[i, 'signal2'] = -1  # 卖出信号
            elif (self.data['close'][i] < self.data['Lower'][i]) and (self.obv[i] > self.obv[i - 1]):
                signals.at[i, 'signal2'] = 1  # 买入信号
        return signals

    def get_strategy_signals(self):
        return self.signals

strategy = BollingerOBVStrategy(df,window=26)
signals = strategy.get_strategy_signals()
dff = signals[["date","close", "Upper"  ,"Lower","OBV","signal","signal2"]]
print(dff[dff['signal2'] != 0])
```

```python
import pandas as pd

class BollingerOBVStrategy:
    def __init__(self, data, window=14, num_std=2):
        self.data = data
        self.window = window
        self.num_std = num_std
        self.obv = self.calculate_obv()
        self.bollinger_bands = self.calculate_bollinger_bands()
        self.signals = self.generate_signals()

    def calculate_obv(self):
        # 计算OBV
        self.data['OBV'] = self.data['volume'] * np.sign(self.data['close'].diff())
        self.data['OBV'] = self.data['OBV'].cumsum().fillna(0)
        return self.data['OBV']

    def calculate_bollinger_bands(self):
        # 计算布林带
        middle_band = self.data['close'].rolling(window=self.window).mean()
        upper_band = middle_band + self.num_std * self.data['close'].rolling(window=self.window).std()
        lower_band = middle_band - self.num_std * self.data['close'].rolling(window=self.window).std()
        self.data['Upper'] = upper_band
        self.data['Middle'] = middle_band
        self.data['Lower'] = lower_band
        return self.data

    def generate_signals(self):
        # 生成交易信号
        signals = self.data.copy()
        signals['signal'] = 0  # 0 for no signal, 1 for buy, -1 for sell

        # 检查OBV和布林带的交叉情况
        for i in range(self.window, len(self.data)):
            if (self.data['close'][i] > self.data['Upper'][i]) and (self.obv[i] < self.obv[i - 1]):
                signals.at[i, 'signal'] = -1  # 卖出信号
            elif (self.data['close'][i] < self.data['Lower'][i]) and (self.obv[i] > self.obv[i - 1]):
                signals.at[i, 'signal'] = 1  # 买入信号
        return signals

    def get_strategy_signals(self):
        return self.signals

strategy = BollingerOBVStrategy(df,window=26)
signals = strategy.get_strategy_signals()
dff = signals[["date","close", "Upper"  ,"Lower","OBV","signal"]]
print(dff[dff['signal'] != 0])
```

```python
import pandas as pd
import pandas_ta as ta

class BollingerOBVStrategy:
    def __init__(self, data, window=14, num_std=2.0):
        self.data = data
        self.window = window
        self.num_std = num_std
        self.data = self.add_ta(self.data)
    def add_ta(self, df):
        # 使用pandas_ta计算OBV和布林带
        df['OBV'] = ta.obv(df['close'], df['volume'])
        df_bb = ta.bbands(df['close'], length=self.window, std=self.num_std)
        df=pd.concat([df, df_bb], axis=1) 
        columns={
            f"BBL_{self.window}_{self.num_std}": "BB_lower",
            f"BBM_{self.window}_{self.num_std}": "BB_middle",
            f"BBU_{self.window}_{self.num_std}": "BB_upper"
        }
        df.rename(columns=columns,inplace=True)
        return df

    def generate_signals(self):
        # 初始化信号列
        df = self.data.copy()
        df['signal'] = 0
        # 生成交易信号：当收盘价突破上轨且OBV下降时可能卖出，突破下轨且OBV上升时可能买入
        df.loc[(df['close'] > df['BB_upper']) & (df['OBV'] <= df['OBV'].shift(1)), "signal"] = -1 
        df.loc[(df['close'] < df['BB_lower']) & (df['OBV'] >= df['OBV'].shift(1)), "signal"] = 1 

        # 过滤信号：只有在OBV确认的情况下才保留信号
        # df['signal'] = df['signal'] * (df['OBV'] > df['OBV'].shift(1))

        return df[['date','close',"BB_upper","BB_lower",'OBV','signal']]
        # return df[['signal']]

    def get_strategy_signals(self):
        if 'BB_upper' not in self.data.columns or 'OBV' not in self.data.columns:
            self.data = self.add_ta(self.data)
        return self.generate_signals()

strategy = BollingerOBVStrategy(df)
signals = strategy.get_strategy_signals()
ss = signals[signals["signal"] != 0]
# print(ss.index)
for ind in ss.index:
    print(signals[ind -5:ind + 5])
```

```python
import pandas as pd
import pandas_ta as ta

class BollingerOBVStrategy:
    def __init__(self, window_bb=5, window_obv=5, window_signal=5):
        self.window_bb = window_bb
        self.window_obv = window_obv
        self.window_signal = window_signal
        self.data = pd.DataFrame(columns=['close', 'volume', 'bb_upper', 'bb_middle', 'bb_lower', 'obv', 'signal'])

    def update_data(self, new_close, new_volume):
        # 添加新数据
        new_data = pd.DataFrame({'close': [new_close], 'volume': [new_volume]})
        self.data = pd.concat([self.data, new_data]) if not self.data.empty else new_data

        # 计算OBV，只有在有足够数据时才计算
        if len(self.data) > 1:
            self.data['obv'] = ta.obv(self.data['close'], self.data['volume'])
        else:
            self.data['obv'] = 0  # 初始化第一行OBV为0

        # 计算布林带和移动平均，只有在有足够的数据时才计算
        if len(self.data) >= self.window_bb:
            # self.data['bb_upper'], self.data['bb_middle'], self.data['bb_lower'],_,L = ta.bbands(self.data['close'], length=self.window_bb)
            # self.data['bb_upper'], self.data['bb_middle'], self.data['bb_lower'],_,L = ta.bbands(self.data['close'], length=self.window_bb)
            df_bb = ta.bbands(self.data['close'], length=self.window_bb)
            columns={
                f"BBL_{self.window_bb}_{2.0}": "bb_lower",
                f"BBM_{self.window_bb}_{2.0}": "bb_middle",
                f"BBU_{self.window_bb}_{2.0}": "bb_upper"
            }
            df_bb.rename(columns=columns,inplace=True)
            self.data=pd.concat([self.data, df_bb], axis=1) 
            self.data['close_ma'] = self.data['close'].rolling(window=self.window_signal).mean()
            self.data['obv_ma'] = self.data['obv'].rolling(window=self.window_signal).mean()

        # 生成信号
        self.generate_signals()

    def generate_signals(self):
        # 清除旧信号
        self.data['signal'] = 0
        # 仅在有足够的数据时生成信号
        if len(self.data) > 1 and len(self.data) >= self.window_signal:
            # 看跌背离：最新收盘价高于前一收盘价，但OBV低于前一OBV的移动平均
            bearish_divergence = self.data.iloc[-1]['close'] > self.data.iloc[-2]['close'] and \
                                  self.data.iloc[-1]['obv'] < self.data.iloc[-2]['obv_ma']

            # 看涨背离：最新收盘价低于前一收盘价，但OBV高于前一OBV的移动平均
            bullish_divergence = self.data.iloc[-1]['close'] < self.data.iloc[-2]['close'] and \
                                  self.data.iloc[-1]['obv'] > self.data.iloc[-2]['obv_ma']

            # 卖出信号：看跌背离且最新收盘价高于布林带上轨
            if bearish_divergence and self.data.iloc[-1]['close'] > self.data.iloc[-1]['bb_upper']:
                self.data.at[len(self.data) - 1, 'signal'] = -1

            # 买入信号：看涨背离且最新收盘价低于布林带下轨
            elif bullish_divergence and self.data.iloc[-1]['close'] < self.data.iloc[-1]['bb_lower']:
                self.data.at[len(self.data) - 1, 'signal'] = 1
    def get_signals(self):
        return self.data['signal']

# 使用示例
strategy = BollingerOBVStrategy()
```

```python
import pandas as pd
import pandas_ta as ta

class BollingerOBVStrategy:
    def __init__(self, window_bb=5, window_obv=5, window_signal=5, num_std=2.0):
        self.window_bb = window_bb
        self.window_obv = window_obv
        self.window_signal = window_signal
        self.num_std = num_std
        self.data = pd.DataFrame(columns=['close', 'volume', 'bb_upper', 'bb_middle', 'bb_lower', 'obv', 'signal'])

    def update_data(self, new_close, new_volume):
        # 如果DataFrame为空，初始化它
        if self.data.empty:
            self.data = pd.DataFrame({'close': [new_close], 'volume': [new_volume]})
            self.data['obv'] = 0  # 初始化OBV
        else:
            # 添加新数据到DataFrame
            new_row = pd.DataFrame({'close': [new_close], 'volume': [new_volume]})
            self.data = pd.concat([self.data, new_row], ignore_index=True)

        # 计算OBV
        if 'obv' in self.data.columns and len(self.data) > 1:
            self.data['obv'] = ta.obv(self.data['close'], self.data['volume'])

        # 计算布林带
        if len(self.data) >= self.window_bb:
            bb_df = ta.bbands(self.data['close'], length=self.window_bb, std=self.num_std)
            columns={
                f"BBL_{self.window_bb}_{self.num_std}": "bb_lower",
                f"BBM_{self.window_bb}_{self.num_std}": "bb_middle",
                f"BBU_{self.window_bb}_{self.num_std}": "bb_upper"
            }
            bb_df.rename(columns=columns,inplace=True)            
            self.data['bb_upper'] = bb_df['bb_upper']
            self.data['bb_middle'] = bb_df['bb_middle']
            self.data['bb_lower'] = bb_df['bb_lower']

        # 生成信号
        self.generate_signals()

    def generate_signals(self):
        # 清除旧信号
        self.data['signal'] = 0

        # 检查背离并生成信号
        if len(self.data) > 1:
            # 看跌背离：价格新高但OBV未新高，且价格超过上轨
            if self.data.iloc[-1]['close'] > self.data.iloc[-2]['close'] and \
               self.data.iloc[-1]['obv'] < self.data.iloc[-2]['obv'] and \
               self.data.iloc[-1]['close'] > self.data.iloc[-1]['bb_upper']:
                self.data.at[len(self.data) - 1, 'signal'] = -1  # 卖出信号

            # 看涨背离：价格新低但OBV未新低，且价格低于下轨
            if self.data.iloc[-1]['close'] < self.data.iloc[-2]['close'] and \
               self.data.iloc[-1]['obv'] > self.data.iloc[-2]['obv'] and \
               self.data.iloc[-1]['close'] < self.data.iloc[-1]['bb_lower']:
                self.data.at[len(self.data) - 1, 'signal'] = 1  # 买入信号

    def get_last_signal(self):
        # 返回最新的信号
        return self.data.iloc[-1]['signal'] if 'signal' in self.data.columns else None

# 使用示例
strategy = BollingerOBVStrategy()
# # 假设实时数据流中新的价格和成交量
# new_close = 150.5
# new_volume = 2000000
# strategy.update_data(new_close, new_volume)
# last_signal = strategy.get_last_signal()
# print(f"Last signal: {last_signal}")
```
