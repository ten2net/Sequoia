```python
import akshare as ak
import pandas as pd
from IPython.display import display

def cal_bid_ask(symbol):
  df = ak.stock_bid_ask_em(symbol=symbol)
  df_buy_volume = df[df['item'].astype(str).str.startswith('buy') & (df['item'].astype(str).str.len() >5)].copy()
  df_sell_volume = df[df['item'].astype(str).str.startswith('sell') & (df['item'].astype(str).str.len() >6)].copy()
  df_buy_volume['value'] = pd.to_numeric(df_buy_volume['value'], errors='coerce')
  df_sell_volume['value'] = pd.to_numeric(df_sell_volume['value'], errors='coerce')
  df_buy_volume=df_buy_volume.head(3)
  df_sell_volume=df_sell_volume.tail(3)
  # print(df_buy_volume)
  # print(df_sell_volume)
  buy = int(df_buy_volume['value'].astype(float, errors='ignore').sum()/100)  
  sell=int(df_sell_volume['value'].astype(float, errors='ignore').sum()/100)
  # print(buy,sell)
  return buy,sell, buy < sell/3

df = ak.stock_hot_rank_em() #当前排名  代码   股票名称   最新价  涨跌幅
df=df[["当前排名","代码","股票名称","最新价","涨跌幅"]]
df.rename(columns={"当前排名":"rank","代码":"code","股票名称":"name","最新价":"price","涨跌幅":"pct"},inplace=True)        
for row in df.values:
  code = (row[1][2:])
  buy,sell,b_lt_s = cal_bid_ask(code)
  if b_lt_s and  buy>0:
    print(code,row[2],buy,sell,b_lt_s )
```

```python
symbols = AmountStockPool().get_symbols(cloumn_name="turnover",k=100)
first_rows_df = pd.DataFrame()
for symbol in symbols:
  stock_df =  ak.stock_zh_a_hist_pre_min_em(symbol=symbol, start_time="09:10:00")
  stock_df=stock_df[stock_df['成交量']>0]
  stock_df['代码'] = symbol
  stock_df['累积成交量'] = stock_df['成交量'].cumsum()
  stock_df['成交量倍数'] = round(stock_df['成交量'] / stock_df['成交量'].shift(-1),2)
  if not stock_df.empty:
    first_row = stock_df.iloc[1:2].copy()  # 使用 iloc[0:1] 来获取第一行的子集
    first_row['代码'] = symbol
    first_rows_df = pd.concat([first_rows_df, first_row], ignore_index=True)  
first_rows_df.sort_values(by="成交量", ascending=False, inplace=True)
print(first_rows_df[["时间","代码","收盘","成交量","成交量倍数"]][:20])
```

```python
from pool.pool import ATPStockPool
df =ATPStockPool(k=200).get_topN()    
stat_data =[]
for ind,row in df.iterrows():
  symbol=row["code"]
  name = row["name"]
  pct = row["pct"]
  turnover = row["turnover"]
  speed5 = row["5_minute_change"]
  amount = round(row["amount"] /(10000 * 10000),2)
  stock_df =  ak.stock_intraday_em(symbol=symbol)
  stock_df=stock_df.query('`时间` >= "09:25:00" and `时间` <= "09:29:00"')
  stat =stock_df.groupby("买卖盘性质")['手数'].sum()
  stat_dict={
    "排名": ind,
    "代码": symbol,
    "名称": name,
    "买盘": stat.get('买盘', 0),
    "卖盘": stat.get('卖盘', 0),
    "中性盘": stat.get('中性盘', 0),
    "总手数": stat.get('买盘', 0) + stat.get('卖盘', 0) +stat.get('中性盘', 0),    
    "买大于卖": stat.get('买盘', 0) >  stat.get('卖盘', 0),
    "涨速":speed5,
    # "换手率":turnover,
    "成交额":amount,
    "涨跌幅": pct
  }
  stat_data.append(stat_dict)
  
stat_data_df = pd.DataFrame(stat_data)  
stat_data_df.sort_values(by="排名", ascending=True, inplace=True)
display(stat_data_df[:30])
```
