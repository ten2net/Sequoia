```python
import akshare as ak
import pandas as pd

def cal_bid_ask(symbol):
  df = ak.stock_bid_ask_em(symbol=symbol)
  df_buy_volume = df[df['item'].astype(str).str.startswith('buy') & (df['item'].astype(str).str.len() >5)].copy()
  df_sell_volume = df[df['item'].astype(str).str.startswith('sell') & (df['item'].astype(str).str.len() >6)].copy()
  df_buy_volume['value'] = pd.to_numeric(df_buy_volume['value'], errors='coerce')
  df_sell_volume['value'] = pd.to_numeric(df_sell_volume['value'], errors='coerce')
  df_buy_volume=df_buy_volume.head(3)
  df_sell_volume=df_sell_volume.tail(3)
  # print(df_buy_volume)
  # print(df_sell_volume)
  buy = int(df_buy_volume['value'].astype(float, errors='ignore').sum()/100)  
  sell=int(df_sell_volume['value'].astype(float, errors='ignore').sum()/100)
  # print(buy,sell)
  return buy,sell, buy < sell/3

df = ak.stock_hot_rank_em() #当前排名  代码   股票名称   最新价  涨跌幅
df=df[["当前排名","代码","股票名称","最新价","涨跌幅"]]
df.rename(columns={"当前排名":"rank","代码":"code","股票名称":"name","最新价":"price","涨跌幅":"pct"},inplace=True)        
for row in df.values:
  code = (row[1][2:])
  buy,sell,b_lt_s = cal_bid_ask(code)
  if b_lt_s and  buy>0:
    print(code,row[2],buy,sell,b_lt_s )
```
