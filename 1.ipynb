```python
import pandas as pd

# 创建示例 DataFrame
df = pd.DataFrame({
    "symbol": ["A", "A", "B", "B", "C", "A"],
    "value": [1, 2, 3, 4, 5, 6],
    # 假设还有其他列
})

# 对每个 'symbol' 组使用 groupby，然后根据 cumcount 过滤，保留每个组的最后一个元素
df_last = df.groupby('symbol').apply(lambda x: x[x.groupby('symbol').cumcount() == (x.groupby('symbol').cumcount().max())])
df_last.reset_index(drop=True, inplace=True)

print(df_last)
```
